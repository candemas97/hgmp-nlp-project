```python
import pandas
import os
```

```python
df_rock = pandas.DataFrame(columns=["Genero", "Artista", "Titulo", "Cancion"])
df_rock
```

```python
for idx, file_name in enumerate(os.listdir("./canciones")):
    artista = file_name.split("-")[0]
    cancion = file_name.split("-")[1].split(".")[0]

    file = open(f"./canciones/{file_name}", "r", encoding="utf-8")
    letra = file.read()
    file.close()

    df_rock.loc[idx, "Genero"] = "rock"
    df_rock.loc[idx, "Artista"] = artista
    df_rock.loc[idx, "Titulo"] = cancion
    df_rock.loc[idx, "Cancion"] = letra
```

```python
df_rock
```

```python
df_rock.to_csv("rock.csv", index=False)
```
